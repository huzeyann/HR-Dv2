```python
# Imports
import torch
torch.cuda.empty_cache()

from utils import *
import numpy as np
from PIL import Image
from sklearn.decomposition import PCA
from sklearn.feature_extraction import grid_to_graph
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, HDBSCAN
import matplotlib.pyplot as plt

from typing import Tuple, List, Any, TypedDict, TypeAlias

ArrList: TypeAlias = List[np.ndarray]

PATCH_SIZE = 14
FEAT_DIM = 384

sofc_temp = Image.open("data/sofc.jpg")
H, W = 490, 490 #sofc_temp.height, sofc_temp.width
IMG_SIZE = 518
PATCH_H, PATCH_W = H // PATCH_SIZE, W // PATCH_SIZE
transform_1 = centre_crop(H, W)
transform_2 = get_input_transform(520, IMG_SIZE)
```

```python
from high_res import HighResDV2

shifts = [1, 2,]
net = HighResDV2("dinov2_vits14", shifts, pattern="Neumann", dtype=torch.float32, pca_each=False)
net.pca_k = 3
net.eval()
net.cuda()

pass
```

We want to examine microstructures and the visualise the features DINOv2 has learned. Also, this serves as an example of how non-square images of different sizes can be used, so long as their dimensions are divisible by the patch size. At the minute sadly the high-res method has a large memory footprint - by definition at least 14x (patch size) that of the low-res model, so large images can't be processed on low budget devices (*i.e,* my laptop's GPU).

```python
sofc_tensor, sofc_img = load_image("data/sofc.jpg", transform_1)
sofc_tensor = sofc_tensor.to("cuda")
sofc_img
```

```python

net.make_high_res(net.dinov2, 14)
low_res_features = net.low_res_features(sofc_tensor.unsqueeze(0))
net.make_high_res(net.dinov2, 4)
net.patch_size = 4
#high_res_features = net.low_res_features(sofc_tensor.unsqueeze(0))
high_res_features, low_res_features = net(sofc_tensor)
```

```python
low_res_features.shape
```

```python
feat_arr_hr, feat_arr_lr = to_numpy(high_res_features), to_numpy(low_res_features)
feat_arr_hr = feat_arr_hr.reshape((net.feat_dim, H * W))
feat_arr_hr = feat_arr_hr.T

fig, axs = plt.subplots(ncols=3, nrows=1)
fig.set_size_inches(20, 12)

axs[0].imshow(sofc_img)
axs[0].set_title("Original", fontsize=24)
axs[0].set_axis_off()

titles = ["DINOv2", "HR-Dv2"]

pcas: List[np.ndarray] = []
features = [feat_arr_lr, feat_arr_hr]
i = 0
for arr, shapes in zip(features, [(120, 120, 3), (490, 490, 3)]): # was (H, W, 3)
    feat_pca = do_single_pca(arr)
    rescale_feat_pca = normalise_pca(feat_pca)
    pcas.append(feat_pca)

    ax = axs[i  + 1]
    ax.imshow(rescale_feat_pca.reshape(shapes)[:, :, :])
    ax.set_title(titles[i], fontsize=24)
    ax.set_axis_off()
    i += 1 
```

```python
reduced_dim_pca = do_single_pca(feat_arr_hr, 70)
reduced_dim_pca = normalise_pca(reduced_dim_pca)
```

```python
clusters = KMeans(n_clusters=2, random_state=0, n_init="auto").fit_predict(reduced_dim_pca)
print(reduced_dim_pca.shape)
#clusters = HDBSCAN(n_jobs=-1).fit_predict(reduced_dim_pca)
#connectivity = grid_to_graph(W, H)
#clusters = AgglomerativeClustering(n_clusters=2, linkage="ward", connectivity=connectivity).fit_predict(reduced_dim_pca)
plt.gcf().set_size_inches(8, 8)
plt.imshow( (clusters.reshape(159, 159, 1)))
```
